## XOR problem in TensorFlow 2.0 (Keras style)

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create dataset with features and labels for use with TensorFlow
x_train, y_train = (tf.constant([[0,0],[0,1],[1,0],[1,1]], "float32"), tf.constant([[0],[1],[1],[0]], "float32"))
```

```python
# Illustration of the problem
XOR_True = [(1, 0), (0, 1)]
XOR_False = [(0, 0), (1, 1)]
feat1, feat2 = zip(*XOR_True)
plt.scatter(feat1, feat2, c="g", label='Output: 1 (XOR = True)')
feat1, feat2 = zip(*XOR_False)
plt.scatter(feat1, feat2, c="r", label='Output: 0 (XOR = False)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.title("Illustration of XOR problem:" "\n" "no straight line can separate the data correctly")
plt.show()
```

### Build and train the model

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(2,)),
  tf.keras.layers.Dense(32, activation=tf.nn.relu),   # hidden layer
  tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)   # output layer
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              #loss='mean_squared_error',   # try this too; treat as regression problem
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=500)
```

```python
print(model.predict(x_train).round())   # since this is a very simple problem with only four input options, 
                                        # there is no separate test set in this case
```

### We can't predict on a real test set here (because there are only four input options), but it would look like this:

```python
y_test = (tf.constant([[0],[1],[1],[0]], "float32"))
```

```python
# Prediction with 100% accuracy
model.evaluate(x_train, y_test)   # this would normally be (x_test, y_test) of course
```
