## The XOR problem: Neural network with 1 hidden layer in Keras

```python
# Import modules 
import numpy as np
import matplotlib as plt
%matplotlib inline

from keras.models import Sequential
from keras.layers.core import Dense
```

```python
# Define features and labels
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[0],[1],[1],[0]], "float32")
```

```python
# MODEL0
model_0 = Sequential()
model_0.add(Dense(4, input_shape=(2,), activation='relu'))   
model_0.add(Dense(1, activation='sigmoid'))
model_0.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy'])
model_0_training = model_0.fit(training_data, target_data, epochs=2000)

print(model_0.predict(training_data).round())
```

```python
#MODEL1
model_1 = Sequential()
model_1.add(Dense(16, input_shape=(2,), activation='relu'))
model_1.add(Dense(1, activation='sigmoid'))
model_1.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy'])
model_1_training = model_1.fit(training_data, target_data, epochs=2000)

print(model_1.predict(training_data).round())
```

```python
# MODEL2
model_2 = Sequential()
model_2.add(Dense(32, input_shape=(2,), activation='relu'))
model_2.add(Dense(1, activation='sigmoid'))
model_2.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy'])
model_2_training = model_2.fit(training_data, target_data, epochs=2000)

print(model_2.predict(training_data).round())
```

```python
# MODEL3 - Like model 2, but without activation function
model_3 = Sequential()
model_3.add(Dense(2, input_shape=(2,), activation='linear'))
model_3.add(Dense(1, activation='sigmoid'))
model_3.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy'])
model_3_training = model_3.fit(training_data, target_data, epochs=2000)

print(model_3.predict(training_data).round())
```
